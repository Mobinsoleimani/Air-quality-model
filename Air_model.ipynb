```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
import seaborn as sns
```

```python
data = pd.read_csv('updated_pollution_dataset.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.isnull().sum()
```

# Encoding_data

```python
encode = preprocessing.LabelEncoder()

data['Air Quality'] = encode.fit_transform(data['Air Quality'])
```

```python
data
```

```python
X = data.drop(columns = ['Air Quality'] , axis = 1)
y = data['Air Quality']
```

```python
X
```

```python
y
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size = 0.3 ,random_state = 42)

```

# Air_Model


```python
model = DecisionTreeClassifier(splitter='random')
model.fit(Xtrain,ytrain)
```

```python
y_hat = model.predict(Xtest)
```

```python
ytest
```

```python
acc = accuracy_score(y_hat,ytest)
```

```python
acc
```

# Confusion Matrix

```python
cm = confusion_matrix(y_hat,ytest)
```

```python
cm
```

```python
print(classification_report(y_hat, ytest))
```
